Contagem de valores de atributos por classe

```python
# https://stackoverflow.com/questions/39132742/groupby-value-counts-on-the-dataframe-pandas
def plot_bar(c1, c2, print_cross_tab, stacked):
    """Plotagem de gráfico de barras usando as colunas c1 e c2. Útil para plotar cada atributo separado por classe."""
    df1 = pd.crosstab(c1, c2)
    if print_cross_tab:
        print()
        print(df1)
        print()
    df1.plot.bar(stacked=stacked)
    
for col in data.select_dtypes(include=['object']):
    plot_bar(data[col], data['Class'], True, False)
```

```python
def plot_hist(df, axes, i, j, pivot_col, field, stacked):
    """Plotagem de histograma com valores separados pela classe"""
    
    axes[i, j].set_title(field)
    # https://stackoverflow.com/questions/41622054/stacked-histogram-of-grouped-values-in-pandas
    df.pivot(columns=pivot_col)[field].plot(kind='hist', stacked=stacked, ax=axes[i, j])

fig, axes = plt.subplots(2, 2, constrained_layout=True, figsize=(12, 6))
cols = data.select_dtypes(include=[np.number]).columns
i_col = 0
for i in range(0, 2):
    for j in range(0, 2):
        plot_hist(data, axes, i, j, 'Class', cols[i_col], True)
        i_col += 1
```
